```python
def colourCode(marketSegment):
    if marketSegment == "OCR":
        return('rgb(255,0,0)')
    elif marketSegment == "RCR":
        return('rgb(0,255,0)')
    else :
        return('rgb(0,0,255)')
```

```python
colourCodeList = [colourCode(x) for x in df_fn['MarketSegment']]
colourCodeList
```
