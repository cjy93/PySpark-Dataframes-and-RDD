```python
'''
Since we know that if we do not limit the the residential to top 20 unsold, we saw some of the other projects have multiple
street names delimited by "/" which makes the exact long lat unsearchable on OneMap. Hence we check which of the street have
a valid long lat and choose the first one that does.

'''
import pandas as pd
# clean my dataset "NoLimit_Unsold_PrivateProperties.csv/part-00000-97b58b7c-7749-4d7e-be35-fffd9bc732bd-c000.csv
df_top20_unsold = pd.read_csv("NoLimit_Unsold_PrivateProperties.csv/part-00000-ee6847a1-c049-473e-bc5e-5c7a61dc6ef6-c000.csv")

#df_top20_unsold is now df of all the units

StreetName = []
for row in df_top20_unsold.iterrows():
    if "/" in row[1][5]:
        a = row[1][5].split("/")
        StreetName.append(a[0])
    else : 
        StreetName.append(row[1][5])
#print(StreetName)

# append this column to df_top20_unsold
df_top20_unsold = df_top20_unsold.assign(Street = StreetName)
print(df_top20_unsold)
```

```python
# make a composite key for both df_top20_unsold
# join df_createdData and df_top20_unsold by key
# https://stackoverflow.com/questions/27928/calculate-distance-between-two-latitude-longitude-points-haversine-formula
df_createdData['combined_key'] = df_createdData['Project']+"_"+df_createdData['Street']
df_top20_unsold['combined_key']= df_top20_unsold['Project']+"_"+df_top20_unsold['Street']
df_top20_unsold
```
