```python
import pandas as pd
import numpy as np
dflonglatestate =  pd.read_csv("createdData/long_lat_district_projects.csv")
dflonglatmrt = pd.read_csv("extra data/mrt_stations_longlat.csv")
print("estate:{}".format(dflonglatestate.head())); print()
print("mrt:{}".format(dflonglatmrt))
```

```python
from math import cos, asin, sqrt
def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295     #Pi/180
    a = 0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2
    return(12742 * asin(sqrt(a))) #2*R*asin...
```

```python
# run this across all the pairs of mrt vs the estates
nameEstate = []
nameStreet = []
nearestMrt = []
mindist = [] 
walkingtime = []
percentagesold =[]
for row in dflonglatestate.iterrows():
    distKMbetwMRT_Estate = []
    for rows in dflonglatmrt.iterrows():
        lat1 = row[1].Latitude
        lon1 = row[1].Longitude
        lon2 = rows[1].Longtitude
        lat2 = rows[1].latitude
        a = distance(lat1,lon1,lat2,lon2)
        
        #print(lat1,lon1,lat2,lon2,a)
        distKMbetwMRT_Estate.append(a) # index is of mrt station
    # https://stackoverflow.com/questions/2474015/getting-the-index-of-the-returned-max-or-min-item-using-max-min-on-a-list
    minindex = distKMbetwMRT_Estate.index(min(distKMbetwMRT_Estate))
    mindist.append(min(distKMbetwMRT_Estate))
    walkingtime.append(min(distKMbetwMRT_Estate)/60)
    nameEstate.append(row[1].Project)
    nameStreet.append(row[1].Street)
    nearestMrt.append(dflonglatmrt["Station Name"][minindex]) # index to find the mrt station
    percentagesold.append(row[1].PercentageSold)
#     print(row[1].Project)
#     print(row[1].Street)
#     print(minindex)
#     print(dflonglatmrt["Station Name"][minindex]) ; print()
```

```python
data = {'project': nameEstate, 'mrt': nearestMrt, 'street':nameStreet,'distToMrt':mindist,'walkingtime':walkingtime,
        'percentagesold': percentagesold}
dfMrtEstateNearest = pd.DataFrame(data=data)
dfMrtEstateNearest

# save to csv
dfMrtEstateNearest.to_csv("createdData/MRT_estate_nearest_dist.csv")
```
